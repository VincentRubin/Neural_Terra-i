```python
def CreateFolder(LocalityPath):
    if not os.path.exists(LocalityPath):
        os.makedirs(LocalityPath)

def GetProductList():
    productList = []
    unzipAllFileArchives(folderPath)
    for folder in os.listdir(folderPath):
        product = ProductIO.readProduct(os.path.join(folderPath,folder, "manifest.safe"))
        productList.append(product)
    return productList

def GetProductListDim(paths):
    
    products = []
    
    for path in paths :
        
        lbs_input_path = os.path.join(path,'*.dim')
        files = glob.glob(lbs_input_path)

        for f in files:
            product = ProductIO.readProduct(f)
            products.append(product)

    return products
```

```python
productOrbits = GetProductListDim(folderOrbits)
productStarts = GetProductListDim(folderResults)

print(len(productOrbits))
print(len(productStarts))
```

```python
geoCodingStart = channelStart.getGeoCoding()
geoCoding = channel.getGeoCoding()

geoPosCurrent = None
pixelPosCurrent = None

startX = 5000
startY = 0
w = 5000
h = 5000

# Create buffer
channelCorrected = array.array('f', (float("NaN") for i in range(0, w * h)))

for y in range(0, h):
    
    clear_output(wait=True)
    print(str(y / h))
    
    for x in range(0, w):
        
        geoPosCurrent = geoCodingStart.getGeoPos(PixelPos(x + startX, y + startY), geoPosCurrent)
        pixelPosCurrent = geoCoding.getPixelPos(geoPosCurrent, pixelPosCurrent)
        
        print(pixelPosCurrent)
        
        # Row Major
        channelCorrected[y * w + x] = channel.getPixelFloat(
            int(pixelPosCurrent.getX()), 
            int(pixelPosCurrent.getY()))
        
print("END")
```

```python
geoCodingStart.dispose()
geoCoding.dispose()
```

```python
# Create new Product
# https://forum.step.esa.int/t/create-a-new-raster-with-new-band/14591/4

fileName = "BRESIL_1_SUBSET_ROTA_2"

product = productOrbits[0]
productNew = Product(fileName, product.getProductType(), w, h)

snappy.ProductUtils.copyGeoCoding(product, productNew)
snappy.ProductUtils.copyMetadata(product, productNew)    

bandNames = product.getBandNames()

productNew.addBand(bandNames[0], ProductData.TYPE_FLOAT32)

writer = ProductIO.getProductWriter('BEAM-DIMAP')
productNew.setProductWriter(writer)
productNew.writeHeader(output_path + "/" + fileName +  ".dim")

bandNew = productNew.getBand(bandNames[0])
bandNew.writePixels(0, 0, w, h, channelCorrected)

productNew.closeIO()
```
